# Simple ML demo with Keras

For more information, see https://keras.io/

```python
# backend can also be set in ~/.keras/keras.json
import os
os.environ["THEANO_FLAGS"] = "floatX=float32,device=gpu"
os.environ["KERAS_BACKEND"] = "tensorflow" # can be "tensorflow" or "theano"
```

```python
from keras.models import Sequential
from keras.layers import Dense
import numpy
```

```python
seed = 7
numpy.random.seed(seed)
```

```python
dataset = numpy.loadtxt("/home/jupyter/jupyter-aws/labs/pima-indians-diabetes.csv", delimiter=",")
```

```python
# split into input (X) and output (Y) variables
X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
# create model
model = Sequential()
model.add(Dense(12, input_dim=8, init='uniform', activation='relu'))
model.add(Dense(8, init='uniform', activation='relu'))
model.add(Dense(1, init='uniform', activation='sigmoid'))
```

```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Fit the model
model.fit(X, Y, nb_epoch=150, batch_size=10)
```

```python
# evaluate the model
scores = model.evaluate(X, Y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
